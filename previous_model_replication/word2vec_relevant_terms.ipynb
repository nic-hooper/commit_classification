```python
import pandas as pd
import numpy as np
import gensim
from gensim.models import Word2Vec
from GensimWord2Vec import GensimWord2VecVectorizer as word2vec
```

```python
messages = pd.read_csv('preprocessed.csv')
messages = messages.drop(labels = ['Unnamed: 0'], axis=1)
messages['clean message'] = messages['Commit message'].apply(lambda x: x.split())
messages.head()
```

```python
extract = messages.loc[messages['Class']=='extract']
extract_w2v = Word2Vec(extract['clean message'],vector_size=750, window=5, epochs=10, hs=1, sg=1)
pd.DataFrame(extract_w2v.wv.most_similar('extract'))
```
